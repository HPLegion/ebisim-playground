```python
import numba as nb
import numpy as np
from ebisim.physconst import MINIMAL_N_1D
```

```python
@nb.njit(cache=True)
def _cubic_spline(x, x1, x2, y1, y2, k1, k2):
    t = (x-x1)/(x2-x1)
    a = k1*(x2-x1) - (y2-y1)
    b = -k2*(x2-x1) + (y2-y1)
    q = (1-t) * y1 + t*y2 + t*(1-t)*((1-t)*a+t*b)
    return q


@nb.njit(cache=True)
def _smooth_to_zero(x):
    N1 = MINIMAL_N_1D
    N2 = 1000*N1
    x = x.copy()
    x[x < N1] = 0
    fil = np.logical_and(N1 < x, x < N2)
    x[fil] = _cubic_spline(x[fil], N1, N2, 0., N2, 0., 1.)
    return x

@nb.njit(cache=True)
def _smooth_temp_rate_to_zero(n, dkT):
    N1 = MINIMAL_N_1D
    N2 = 1000*N1
    dkT = dkT.copy()
    dkT[n < N1] = 0
    fil = np.logical_and(N1 < n, n < N2)
    dkT[fil] = dkT[fil] * _cubic_spline(n[fil], N1, N2, 0., 1., 0., 1.)
    return dkT
```

```python
n = np.geomspace(1e-7, 1e-2,1000)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.loglog(n, _smooth_to_zero(n))
```

```python
plt.loglog(n, _smooth_temp_rate_to_zero(n, np.ones_like(n)*1000000e-3))
```
