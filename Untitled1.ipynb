```python
import ebisim as eb
import numpy as np
import matplotlib.pyplot as plt
plt.rc("figure", figsize=(7.5,6), dpi=150)
```

```python
R_E = 100e-6
R_DT = 7e-3
V_AX = 500
CUR = 170e-3
E_KIN = 1900
LEN = 80e-3
B_0 = 5.5
```

```python
o = eb.Target.get_gas("O",3e-11, R_DT,cx=False) #4Kelvin?
h2 = eb.Target.get_gas("H",2e-9, R_DT,cx=False) #4Kelvin?
h = eb.Target.get_ions("H", 1e9, kT=30)
```

```python
dev = eb.Device.get(CUR, E_KIN, R_E, LEN, V_AX, B_0, R_DT, r_dt_bar=3e-3, n_grid=420)
```

```python
h2res,  ores= eb.advanced_simulation(dev, [h2, o],5, n_threads=4, rates=True, options=eb.ModelOptions(RADIAL_DYNAMICS=True, RECOMPUTE_CROSS_SECTIONS=True))
```

```python
_=ores.plot_rate(eb.Rate.E_KIN_MEAN, xscale="linear", ylim=(0,2500))
# ores.plot_rate(eb.Rate.V_AX)
# ores.plot_rate(eb.Rate.V_RA)
```

```python
ores.plot(xscale="linear")
# hres.plot(xscale="linear")
_=h2res.plot(xscale="linear")
```
